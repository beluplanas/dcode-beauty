```python
import requests
from bs4 import BeautifulSoup
import time 
import random
import pandas as pd
```

```python
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}
URL_BASE = 'https://cosmetics.specialchem.com'
links = []
```

```python
for i in range(1,158):
    print(f'scraping page {i}')
    url = f'https://cosmetics.specialchem.com/selectors?indexpage={i}'
    r = requests.get(url, headers=header)
    print(r.status_code)
    soup = BeautifulSoup(r.content, "html.parser")
    items = soup.find_all("div", class_="box_product")
    for item in items:
        val = item.find('a',href=True)
        links.append(URL_BASE+val['href'])
len(links)
```

```python
#guardar en csv links para scrapear descripcion
pd.DataFrame(links).to_csv('links_to_scrape.csv')
```
